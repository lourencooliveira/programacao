# Lista de 100 Exercícios de Python

Seja bem-vindo(a) a esta trilha prática de exercícios em Python.

Esta atividade foi organizada com o objetivo de fortalecer sua aprendizagem em lógica de programação, desenvolvimento de algoritmos e resolução de problemas computacionais por meio da prática progressiva. Ao longo deste material, você deverá resolver os **100 exercícios propostos**, estruturados em nível crescente de complexidade, contemplando desde conceitos introdutórios até o uso de listas, funções e dicionários.

Para a realização da atividade, é fundamental que cada solução seja desenvolvida com atenção à **clareza, organização e legibilidade do código**.

Ao finalizar a resolução dos exercícios, salve seus códigos em um **repositório no GitHub** e envie o link no Moodle.
